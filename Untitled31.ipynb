# **PistonTheta: Efecto Casimir en aislantes topológicos**

```python
import numpy as np
from scipy.special import roots_legendre
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

# Constantes

```python
ALPHA_FINE   = 1.0 / 137.0   # constante de estructura fina
OMEGA_E_NORM = 0.3         # ωₑ/ωᵣ (plasma del TI normalizado)
GAMMA_NORM   = 0.01          # γ/ωᵣ (amortiguamiento normalizado)
```

# Pistón θ

```python

class PistonTheta:


    #Energía y fuerza de referencia (EN EL LÍMITE DE CONDUCTORES PERFECTOS)
    EL = -np.pi**2 / 720.0
    FL = -np.pi**2 / 240.0

    def __init__(self,
                 theta      = np.pi,
                 D_over_L   = 0.01,
                 alpha      = ALPHA_FINE,
                 omega_e    = OMEGA_E_NORM,
                 gamma      = GAMMA_NORM,
                 n_rho      = 120,
                 n_phi      = 32,
                 rho_max    = 35.0):

        self.theta    = theta
        self.D_over_L = D_over_L
        self.alpha    = alpha
        self.omega_e  = omega_e
        self.gamma    = gamma
        self.n_rho    = n_rho
        self.n_phi    = n_phi
        self.rho_max  = rho_max

        # Pre-calcular nodos de Gauss-Legendre
        self._gl_nodes, self._gl_weights = roots_legendre(max(n_rho, n_phi))

    def __repr__(self):
        return (f"PistonTheta(θ={self.theta/np.pi:.2f}π, "
                f"D/L={self.D_over_L}, "
                f"n_rho={self.n_rho}, n_phi={self.n_phi})")


    #  Permitividad dieléctrica en frecuencias de Matsubara

    def epsilon(self, xi):
        xi = max(float(xi), 0.0)
        return 1.0 + self.omega_e**2 / (xi**2 + 1.0 + self.gamma * xi)


    #  MATRICES DE REFLEXIÓN DE UNA INTERFAZ θ
    def _RT_interface(self, xi, k, n1, n2, sign_grad):
        """
        Matrices de reflexión de una interfaz θ.

        Retorna (R, R_tilde):
          R      : reflexión vista desde n1
          R_tilde: reflexión vista desde n2
        """
        xi = max(float(xi), 1e-14)
        k  = max(float(k),  1e-14)

        ab_bar = self.alpha * self.theta / np.pi  # ᾱ = α|θ|/π

        a = xi**2 + k**2 / n1**2
        b = xi**2 + k**2 / n2**2

        sqrt_ab = np.sqrt(a * b)
        if not np.isfinite(sqrt_ab) or sqrt_ab < 1e-300:
            return np.zeros((2, 2)), np.zeros((2, 2))

        chi_plus  = (a + b) / sqrt_ab
        chi_minus = (a - b) / sqrt_ab
        ab2       = ab_bar**2

        Delta = n1**2 + n2**2 + ab2 + n1 * n2 * chi_plus
        if not np.isfinite(Delta) or abs(Delta) < 1e-300:
            return np.zeros((2, 2)), np.zeros((2, 2))

        # Reflexión desde n1
        r_TE  = ( n1**2 - n2**2 - ab2 + n1*n2*chi_minus) / Delta
        r_TM  = (-n1**2 + n2**2 + ab2 + n1*n2*chi_minus) / Delta
        r_off = sign_grad * 2.0 * ab_bar * n1 / Delta

        R = np.array([[r_TE,  r_off],
                      [r_off, r_TM ]])

        # Reflexión desde n2 (interfaz invertida)
        r_TE_t  = ( n2**2 - n1**2 - ab2 - n1*n2*chi_minus) / Delta
        r_TM_t  = (-n2**2 + n1**2 + ab2 - n1*n2*chi_minus) / Delta
        r_off_t = -sign_grad * 2.0 * ab_bar * n2 / Delta

        R_til = np.array([[r_TE_t,  r_off_t],
                          [r_off_t, r_TM_t ]])

        return R, R_til


    #COMPOSICIÓN DE ESPEJOS (fórmula de Airy matricial) Reflexión efectiva del stack {R_far | propagador(p²) | R_near} vista desde el exterior (lado de R_near). Usa la relación de Stokes correcta en Matsubara: T·T̃ = I - R·R̃
    @staticmethod
    def _airy_combine(R_near, R_near_til, p2, R_far):

        I        = np.eye(2)
        TT_til   = I - R_near @ R_near_til
        denom    = I - p2 * R_far @ R_near_til
        try:
            inv_d = np.linalg.inv(denom)
        except np.linalg.LinAlgError:
            return R_near.copy()
        return R_near + TT_til @ (p2 * inv_d @ R_far)


    #  KERNEL DEL INTEGRANDO
    def _kernel(self, rho, phi, chi):
        """
        Integrando de Casimir en coordenadas polares (ρ, φ):
            ρ²·sin(φ) · ln det[I - M(ρ,φ)]

        donde M es la matriz de round-trip dentro del bloque TI.
        """
        xi = max(rho * np.cos(phi), 1e-14)
        k  = max(rho * np.sin(phi), 1e-14)

        n_vac = 1.0
        n_TI  = np.sqrt(self.epsilon(xi))

        # Momentos normales evanescentes (Matsubara)
        kappa_vac = rho
        kappa_TI  = np.sqrt(xi**2 + k**2 / n_TI**2)

        # Propagadores de doble pasada
        p2_L = np.exp(-2.0 * kappa_vac * chi)
        p2_D = np.exp(-2.0 * kappa_TI  * self.D_over_L)
        p2_R = np.exp(-2.0 * kappa_vac * (1.0 - chi - self.D_over_L))

        if not all(np.isfinite(x) for x in [p2_L, p2_D, p2_R]):
            return 0.0

        R_PEC = np.array([[-1., 0.], [0., 1.]])

        # Matrices de reflexión de cada interfaz
        RS1, RS1_til = self._RT_interface(xi, k, n_vac, n_TI, sign_grad=+1)
        RS2, RS2_til = self._RT_interface(xi, k, n_TI, n_vac, sign_grad=-1)

        # Reflexión efectiva izquierda (vista desde el TI): {PEC_L + vac_L + Σ₁}
        R_left = self._airy_combine(RS1_til, RS1, p2_L, R_PEC)

        # Reflexión efectiva derecha (vista desde el TI): {Σ₂ + vac_R + PEC_R}
        R_right = self._airy_combine(RS2, RS2_til, p2_R, R_PEC)

        # Round-trip dentro del TI
        M = p2_D * (R_right @ R_left)

        sign, logdet = np.linalg.slogdet(np.eye(2) - M)
        if sign <= 0 or not np.isfinite(logdet):
            return 0.0

        return rho**2 * np.sin(phi) * logdet


    #  INTEGRACIÓN
    def _integrate(self, chi):
        """Integral 2D en (ρ, φ) para un χ dado."""
        # Cutoff adaptativo
        w_min    = max(min(chi, self.D_over_L, 1.0 - chi - self.D_over_L), 0.01)
        rho_cut  = min(self.rho_max, 30.0 / w_min)

        nodes_rho = 0.5 * (1.0 + self._gl_nodes[:self.n_rho]) * rho_cut
        wgts_rho  = self._gl_weights[:self.n_rho] * (rho_cut / 2.0)

        nodes_phi = 0.5 * (1.0 + self._gl_nodes[:self.n_phi]) * (np.pi / 2.0)
        wgts_phi  = self._gl_weights[:self.n_phi] * (np.pi / 4.0)

        total = 0.0
        for rho, wr in zip(nodes_rho, wgts_rho):
            inner = sum(wp * self._kernel(rho, phi, chi)
                        for phi, wp in zip(nodes_phi, wgts_phi))
            total += wr * inner

        return total / (4.0 * np.pi**2)

    #Malla de la integración
    def energia(self, chi):

        if not (0.0 < chi < 1.0 - self.D_over_L):
            raise ValueError(f"χ debe estar en (0, {1-self.D_over_L:.3f})")
        return self._integrate(chi)

    def energia_topologica(self, chi):
        """
        Contribución topológica Eθ = E(χ) - EL.
        """
        return self.energia(chi) - self.EL

    def energia_topologica_norm(self, chi):
        """
        Ratio Eθ/EL.
        """
        return self.energia_topologica(chi) / self.EL

    def fuerza(self, chi, dchi=0.008):
        """
        Fuerza sobre el pistón F = -dE/d(x₀) = -(1/L)·dE/dχ
        calculada por diferencias centradas.

        Parámetros
        ----------
        chi  : float  — posición relativa χ = x₀/L
        dchi : float  — paso para diferencias centradas (default: 0.008)
        """
        chi_p = chi + dchi
        chi_m = chi - dchi
        if chi_p >= 1.0 - self.D_over_L - 0.005:
            raise ValueError("chi + dchi demasiado cerca del borde derecho")
        if chi_m <= 0.005:
            raise ValueError("chi - dchi demasiado cerca del borde izquierdo")
        return -(self.energia(chi_p) - self.energia(chi_m)) / (2.0 * dchi)

    def fuerza_norm(self, chi, dchi=0.008):
        """Ratio Fθp/FL."""
        return self.fuerza(chi, dchi) / self.FL


    #  VERIFICACIÓN DE LÍMITES

    def verificar(self, chi=0.5):
        """
        Verifica dos límites analíticos conocidos:

        1. θ=0, D→0  →  E/EL ≈ 1  (cavidad PEC pura)
        2. θ→∞       →  tres cavidades PEC independientes
        """
        print(f"{'='*55}")
        print(f"  VERIFICACIÓN — θ={self.theta/np.pi:.1f}π, D/L={self.D_over_L}, χ={chi}")
        print(f"{'='*55}")

        # Test 1: límite θ=0
        p0 = PistonTheta(theta=0.0, D_over_L=0.01,
                         n_rho=self.n_rho, n_phi=self.n_phi)
        E0 = p0.energia(chi)
        print(f"\nTest 1: θ=0, D/L=0.01, χ={chi}")
        print(f"  Esperado : E/EL = 1.0000")
        print(f"  Obtenido : E/EL = {E0/self.EL:.5f}")

        # Test 2: límite θ→∞
        D, c = self.D_over_L, chi
        E_analitico = self.EL * (c**(-3) + D**(-3) + (1-c-D)**(-3))
        p_inf = PistonTheta(theta=1e5*np.pi, D_over_L=D,
                            n_rho=self.n_rho, n_phi=self.n_phi)
        E_inf = p_inf.energia(chi)
        print(f"\nTest 2: θ→∞, D/L={D}, χ={chi}")
        print(f"  Esperado (3 cavidades): E = {E_analitico:.4e}")
        print(f"  Obtenido              : E = {E_inf:.4e}")
        print(f"  Ratio                 : {E_inf/E_analitico:.5f}  (→ 1)")
        print()


    #  GRÁFICAS

    def plot_energia(self,
                     chi_vals   = None,
                     theta_list = None,
                     ax         = None,
                     save       = None):

        if chi_vals is None:
            chi_vals = _chi_grid(self.D_over_L)

        if theta_list is None:
            theta_list = [self.theta]

        show = ax is None
        if show:
            fig, ax = plt.subplots(figsize=(8, 6))

        colors = plt.cm.plasma(np.linspace(0.15, 0.85, len(theta_list)))

        for theta, color in zip(theta_list, colors):
            p = PistonTheta(theta=theta, D_over_L=self.D_over_L,
                            n_rho=self.n_rho, n_phi=self.n_phi)
            ratios = []
            for chi in chi_vals:
                try:
                    ratios.append(p.energia_topologica_norm(chi))
                except ValueError:
                    ratios.append(np.nan)

            ax.plot(chi_vals, ratios, marker='o', markersize=3,
                    linewidth=1.3, color=color,
                    label=fr"$\theta={theta/np.pi:.0f}\pi$")

        ax.set_xlabel(r"$\chi = x_0/L$", fontsize=13)
        ax.set_ylabel(r"$\mathcal{E}_\theta\,/\,\mathcal{E}_L$", fontsize=13)
        ax.set_title(fr"Energía Casimir — D/L={self.D_over_L}", fontsize=12)
        ax.set_xlim(0, 1)
        ax.grid(True, alpha=0.35)
        ax.legend(fontsize=11)

        if save:
            plt.tight_layout()
            plt.savefig(save, dpi=150)
            print(f"Guardado: {save}")
        if show:
            plt.tight_layout()
            plt.show()

    def plot_fuerza(self,
                    chi_vals   = None,
                    theta_list = None,
                    dchi       = 0.008,
                    ax         = None,
                    save       = None):

        if chi_vals is None:
            chi_vals = _chi_grid(self.D_over_L, margen=0.04)

        if theta_list is None:
            theta_list = [self.theta]

        show = ax is None
        if show:
            fig, ax = plt.subplots(figsize=(8, 6))

        colors = plt.cm.plasma(np.linspace(0.15, 0.85, len(theta_list)))

        for theta, color in zip(theta_list, colors):
            p = PistonTheta(theta=theta, D_over_L=self.D_over_L,
                            n_rho=self.n_rho, n_phi=self.n_phi)
            forces = []
            for chi in chi_vals:
                try:
                    forces.append(p.fuerza_norm(chi, dchi))
                except ValueError:
                    forces.append(np.nan)

            ax.plot(chi_vals, forces, marker='o', markersize=3,
                    linewidth=1.3, color=color,
                    label=fr"$\theta={theta/np.pi:.0f}\pi$")

        ax.axhline(0, color='k', linewidth=0.8, linestyle='--', alpha=0.6)
        ax.set_xlabel(r"$\chi = x_0/L$", fontsize=13)
        ax.set_ylabel(r"$F_{\theta p}\,/\,F_L$", fontsize=13)
        ax.set_title(fr"Fuerza Casimir — D/L={self.D_over_L}", fontsize=12)
        ax.set_xlim(0, 1)
        ax.set_ylim(-1.2, 1.2)
        ax.grid(True, alpha=0.35)
        ax.legend(fontsize=11)

        if save:
            plt.tight_layout()
            plt.savefig(save, dpi=150)
            print(f"Guardado: {save}")
        if show:
            plt.tight_layout()
            plt.show()

    def scan(self, chi_vals=None, verbose=True):

        if chi_vals is None:
            chi_vals = _chi_grid(self.D_over_L)

        resultados = {'chi': [], 'E': [], 'E_theta': [],
                      'E_norm': [], 'F': [], 'F_norm': []}

        for chi in chi_vals:
            try:
                E  = self.energia(chi)
                Et = E - self.EL
                En = Et / self.EL
                try:
                    F  = self.fuerza(chi)
                    Fn = F / self.FL
                except ValueError:
                    F = Fn = np.nan
            except ValueError:
                E = Et = En = F = Fn = np.nan

            resultados['chi'].append(chi)
            resultados['E'].append(E)
            resultados['E_theta'].append(Et)
            resultados['E_norm'].append(En)
            resultados['F'].append(F)
            resultados['F_norm'].append(Fn)

            if verbose:
                print(f"  χ={chi:.3f}  E/EL={En:+.5f}  F/FL={Fn:+.5f}")

        return {k: np.array(v) for k, v in resultados.items()}
```

Malla de integración

```python
def _chi_grid(D_over_L, margen=0.02, n_centro=18, n_borde=6):
    """Malla de χ densa en los bordes, suave en el centro."""
    return np.concatenate([
        np.linspace(margen,        0.15,          n_borde, endpoint=False),
        np.linspace(0.15,          1-D_over_L-0.15, n_centro),
        np.linspace(1-D_over_L-0.15, 1-D_over_L-margen, n_borde)
    ])
```

```python
def comparar_thetas(theta_list, D_over_L=0.01, n_rho=120, n_phi=32,
                    guardar_prefijo=None):
   # Usamos la primera instancia como base para las gráficas
    p = PistonTheta(theta=theta_list[0], D_over_L=D_over_L,
                    n_rho=n_rho, n_phi=n_phi)

    save_e = f"{guardar_prefijo}_energia.png" if guardar_prefijo else None
    save_f = f"{guardar_prefijo}_fuerza.png"  if guardar_prefijo else None

    p.plot_energia(theta_list=theta_list, save=save_e)
    p.plot_fuerza( theta_list=theta_list, save=save_f)
```

```python
if __name__ == "__main__":

    # ── 1. Uso básico ─────────────────────────────────────────────────────────
    p = PistonTheta(theta=7*np.pi, D_over_L=0.01)
    print(p)

    E  = p.energia(chi=0.3)
    En = p.energia_topologica_norm(chi=0.3)
    F  = p.fuerza(chi=0.3)
    Fn = p.fuerza_norm(chi=0.3)
    print(f"\nχ=0.3:  E={E:.4e}  Eθ/EL={En:+.5f}  F={F:.4e}  Fθp/FL={Fn:+.5f}")

    # ── 2. Verificar límites ──────────────────────────────────────────────────
    p.verificar(chi=0.5)

    # ── 3. Scan completo ──────────────────────────────────────────────────────
    print("\nScan completo χ ∈ (0,1):")
    datos = p.scan()

    # ── 4. Reproducir Fig. 2 del paper ───────────────────────────────────────
    print("\nEθ/EL vs χ ...")
    comparar_thetas(
        theta_list=[3*np.pi, 5*np.pi, 7*np.pi, 9*np.pi, 11*np.pi],
        D_over_L=0.01,
        )

    # ── 5. Reproducir Fig. 3 del paper ───────────────────────────────────────
    print("\n Fθ/FL vs χ ...")
    comparar_thetas(
        theta_list=[7*np.pi, 15*np.pi, 23*np.pi, 31*np.pi, 39*np.pi],
        D_over_L=0.01,
        )

    # ── 6. Explorar efecto del grosor D ──────────────────────────────────────
    print("\nEfecto del grosor del TI (D/L):")
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    for ax, D in zip(axes, [0.005, 0.01, 0.05]):
        p_D = PistonTheta(theta=7*np.pi, D_over_L=D)
        p_D.plot_energia(
            theta_list=[3*np.pi, 7*np.pi, 11*np.pi],
            ax=ax
        )
        ax.set_title(f"D/L = {D}")
    plt.tight_layout()
    plt.savefig("god plot", dpi=150)
    plt.show()
```
